## Verifying how home teams performed through the years, calculating home team win percentage 

match_result is a column that will contain information about the match result
<br>
1- Home team win
<br>
2- Visitor team win
<br>
3- Draw

```python
match_result = []

for i, r in first_division.iterrows():
    if r["local_goals"] > r["visitor_goals"]:
        match_result.append(1)
    elif r["local_goals"] < r["visitor_goals"]:
        match_result.append(2)
    else:
        match_result.append(3)
```

```python
first_division["match_result"] = pd.Series(match_result)
```

```python
first_division.head(10)
```

Two dictionaries are created, one will store the total of home wins for each season, the other one will store the total of games for each season

```python
total_home_wins_season = {}
total_games_season = {}

for i, r in first_division.iterrows():
    if r["season"] not in total_home_wins_season:
        total_games_season[r["season"]] = 1
    else:
        total_games_season[r["season"]] += 1
    
    if r["match_result"] == 1:
        if r["season"] not in total_home_wins_season:
            total_home_wins_season[r["season"]] = 1
        else:
            total_home_wins_season[r["season"]] += 1
```

pct_home_wins_pct will store the percentage of home team wins for each season

```python
pct_home_wins_pct = {}

for r in total_home_wins_season:
    pct_home_wins_pct[r] = total_home_wins_season[r] / total_games_season[r]
```

```python
pct_home_wins_pct = pd.Series(pct_home_wins_pct)
```

```python
pct_home_wins_pct.sort_values()
```

```python
ax = plt.gca()

pd.DataFrame(pct_home_wins_pct).plot(kind='line', ax=ax)
plt.show()
```

Analysing the graph is possible to sset that in the 70s the home win pct were really high
<br>
However, there is a sharp drop in the 84-85 season, but the 05-06 has the lowest home team win percentage.

## Analysing 2005-2006 season

Selecting only the matches from 05-06 season

```python
season_05_06 = first_division[first_division["season"] == "2005-06"].copy()
season_05_06.reset_index(inplace=True)
```

Dropping columns that are not necessaring

```python
season_05_06.drop(["level_0", "index", "id"], axis=1, inplace=True)
```

```python
season_05_06.head()
```

This low home win percentage can be explained with two hypothesis
<br>
1- One team had a high visitor win percentage
<br>
2- One team had a low home win percentage

```python
def calculate_wins(df, result):
    """
    The function receives to parameters, a dataframe and the 
    result that will be calculated.
    1 - Home wins
    2 - Visitor wins
    A pandas Series is returned with the team name and its number of victories
    """
    
    if result == 1:
        team = "local_team"
    elif result == 2:
        team = "visitor_team"
    else:
        print("N√£o √© possivel calcular o n√∫mero de vit√≥rias com o resultado empate!")
        return
    
    
    result_dict = {}
    
    for i, r in df.iterrows():
        if r["match_result"] == result:
            if r[team] not in result_dict:
                result_dict[r[team]] = 1
            else:
                result_dict[r[team]] += 1
    return pd.Series(result_dict)
            
```

Calculating the number and percentage of home and away wins for each team

```python
away_wins_team = calculate_wins(season_05_06,2)
```

```python
away_wins_team.sort_values()
```

```python
home_wins_team = calculate_wins(season_05_06,1)
```

```python
home_wins_team.sort_values()
```

```python
pct_home_wins_team = ((home_wins_team)/((len(home_wins_team.index.unique())-1)*2))*100
```

```python
pct_home_wins_team.sort_values()
```

```python
pct_away_wins_team = ((away_wins_team)/((len(away_wins_team.index.unique())-1)*2))*100
```

```python
pct_away_wins_team.sort_values()
```

Creating a dataframe that  contain the number and the percentage of home and away wins for each team

```python
wins_team_05_06 = pd.concat([home_wins_team, pct_home_wins_team, away_wins_team, pct_away_wins_team], axis=1, sort=False)
wins_team_05_06.columns = ["home_wins_team", "pct_home_wins_team", "away_wins_team", "pct_away_wins_team"]
wins_team_05_06.sort_index(inplace=True)
```

We can see in the graph that the difference between the lowest and the highest number of home victories is much bigger than visitor wins

Teams like Malaga, Cadiz and Santander had less than 6 victories as home team.

The conclusion is that the lowest historic home win percentage is explained by the poor performance of some teams
